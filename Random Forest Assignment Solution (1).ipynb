```python
## checking for outliers Manually
for i in df.loc[:,features.columns]:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    up=Q3+1.5*IQR
    low=Q1-1.5*IQR
    
    if df[(df[i]>=up) | (df[i]<=low)].any(axis=None):
        print(i,'yes')
    else:
        print(i,'No')
    
    
```

```python
import seaborn as sns
```

```python
## checking outliers for each column seperately
k=1
plt.figure(figsize=(12,10))
plt.suptitle('Distrubation of Outliers')
for i in df.loc[:,features.columns]:
    plt.subplot(5,2,k)
    sns.boxplot(x=i,data=df.loc[:,features.columns])
    plt.tight_layout()
    k+=1
    plt.show()
    

```

```python
## Removing outliers
def remove_outliers(df,column):
    for column in features.columns:
        Q1=df[column].quantile(0.25)
        Q3=df[column].quantile(0.75)
        IQR=Q3-Q1
        low=Q1-1.5*IQR
        up=Q3+1.5*IQR
    return df[(df[column] >= low) & (df[column] <= up)]

    
```

```python
df_cleaned=remove_outliers(df,features.columns)
```

```python
df_cleaned
```

```python
## clearly outliers has been removed since earlier no of rows was 213 but now it is 201.
```

## 4: Random Forest Model Implementation

```python
df_cleaned.Type.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features1,target,train_size=0.80,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=50,max_depth=3,max_features=1,bootstrap=False,random_state=100)
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
