## Task 2: Dimensionality Reduction with PCA:

```python
## standardisation of data
from sklearn.preprocessing import StandardScaler
```

```python
std_sca=StandardScaler()
```

```python
new_feat=pd.DataFrame(std_sca.fit_transform(features),columns=features.columns)
new_feat
```

```python
## checking for mean and std dev.
import numpy as np
```

```python
new_feat.describe()
```

```python
 # Implementing  PCA to reduce the dimensionality of the dataset
```

```python
new_feat.T
```

```python
cov=np.cov(new_feat.T)
```

```python
eigen_values,eigen_vectors=np.linalg.eig(cov)
eigen_values
```

```python
figure=plt.figure(figsize=(10,6))
sing_vals=np.arange(len(eigen_values)) + 1
plt.plot(sing_vals,eigen_values, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue') 
plt.show() 
```

```python
## since curve is decreasing   high for the value of principal component having 2,3,4 so we can take optimum value for PCA either 2 or 3 or 4. 
```

```python
## Transforming original dataset into PCA.
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=4)
```

```python
new_feat_pca=pca.fit(new_feat)
```

```python
## new_feat_pca=features after applying PCA in new_feat
new_feat_pca=pd.DataFrame(pca.transform(new_feat),columns=['A','B','C','D'])
new_feat_pca
```

## Task 3: Clustering with Original Data:

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# new_feat = oringinal features after standardisation
new_feat.head()
```

```python
## now to define KMeans we need optimum value of clusters for this we do the following.
```

```python
inertia=[]
for i in range(1,50):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(new_feat)
    inertia.append(kmeans.inertia_)
print(inertia)
```

```python
## plotting Elbow curve
plt.plot(range(1,50),inertia)
plt.xlabel('Num of clusters')
plt.ylabel('inertia')
plt.title('Elbow mehod',size=20,color='orange',y=(1.1))
plt.show()
```

```python
kmeans=KMeans(n_clusters=20)
```

```python
cluster=kmeans.fit(new_feat)
```

```python
cluster.labels_
```

```python
np.unique(cluster.labels_)
```

```python
# visualisation of clustering result
new_feat['cluster']=cluster.labels_
```

```python
new_feat.head()
```

```python
import seaborn as sns 
```

```python
sns.scatterplot(data=new_feat,x='Phenols',y='Proanthocyanins',hue='cluster')
plt.show()
```

```python
## Evaluation of clustering perfomance
```

```python
score=silhouette_score(new_feat,cluster.labels_)
score
```
