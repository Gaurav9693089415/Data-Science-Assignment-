## Data Preprocessing 

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('Online retail.xlsx',header=None,)
```

```python
df.shape
```

```python
df.head()
```

```python
dg = df[0].str.split(',') # if we will use expand=True then we will get it as dataframe.But here dataframe is not required when we are applying TransactionnEncoder.
dg.head()
```

```python
!pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
tra_enc=TransactionEncoder()
```

```python
tra=tra_enc.fit_transform(dg)
tra
```

```python
arr=tra.astype('int')
arr
```

```python
arr.shape
```

```python
data=pd.DataFrame(arr,columns=tra_enc.columns_)
data
```

```python
print(tra_enc.columns_)
```

## Association rule mining

```python
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings('ignore')
```

```python
# when we use min_support=0.5 it is not working i.e it has no support value grater than 0.5 so we will get nothing.and we use min_support=0.01


freq_items=apriori(data,min_support=0.01,use_colnames=True)# when we use min_support=0.5 it is not working i.e it has no support value grater than 0.5 so we will get nothing.and we use min_support=0.01
freq_items
```

```python
## when we use imn_threshold=0.5 or greater than that we get nothing so  to avoid that situation we used min_threshold=0.2
association_rules(freq_items,min_threshold=0.2,metric='confidence')
```
