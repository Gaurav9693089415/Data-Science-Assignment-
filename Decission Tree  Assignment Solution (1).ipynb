## 4. Decision Tree Classification:

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features1,target,train_size=0.80,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec_tr_cls=DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=3)
```

```python
dec_tr_cls.fit(x_train,y_train)
```

```python
y_pred=dec_tr_cls.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
Label encoding and one-hot encoding are two common techniques used to convert categorical data into numerical format for machine learning models. Here’s how they differ:

Label Encoding
Description: Label encoding assigns a unique integer to each category in a feature.
Example: For a feature “Color” with categories [“Red”, “Green”, “Blue”], label encoding might assign [0, 1, 2] respectively.
Use Case: Suitable for ordinal data where the categories have a meaningful order (e.g., “Low”, “Medium”, “High”).
Pros: Simple and memory-efficient.
Cons: Can introduce ordinal relationships where none exist, potentially misleading the model.
```
