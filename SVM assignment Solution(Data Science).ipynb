```python
## since we have many categorical columns so we can't find correlation for our dataset.so  first we have to encode the categorical columns.
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc=LabelEncoder()
```

```python
for column in df.select_dtypes(include=['object']).columns:
    df[column]=lab_enc.fit_transform(df[column])
                                                              ## Clearly data has been encoded
    
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features=df.drop('class',axis=1)
```

```python
features.head()
```

```python
target=df['class']
target.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,random_state=100,train_size=0.75)
```

```python
print('x_train:',x_train.shape)
print('x_test:',x_test.shape)
print('y_train:',y_train.shape)
print('y_test:',y_test.shape)
```

## SVM Implementation

```python
Task 4: SVM Implementation
1.	Implement a basic SVM classifier using Python libraries like scikit-learn.
2.	Train the SVM model on the training data.
3.	Evaluate model performance on the testing data using appropriate metrics (e.g., accuracy, precision, recall, F1-score).
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='rbf')
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
pd.DataFrame({'y_test':y_test,'y_pred':y_pred})
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

## Task 5: Visualization of SVM Results
1.	Visualize classification results on the testing data


```python
## Visualisation of results
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## Task 6: Parameter Tuning and Optimization
1.	Experiment with different SVM hyperparameters (e.g., kernel type, regularization parameter) to optimize performance.


```python
from sklearn.model_selection import GridSearchCV
```

```python
kernel={'kernel':['linear','rbf','poly']}
```

```python
grid=GridSearchCV(SVC(),kernel,refit=True,verbose=2)
```

```python
grid.fit(x_train,y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

## Task 7: Comparison and Analysis
1.	Compare SVM performance with various kernels (e.g., linear, polynomial, radial basis function).


```python
kernel=['linear','rbf','poly']

for i in kernel:
    svc=SVC(kernel=i)
    svc.fit(x_train,y_train)
    y_pred=svc.predict(x_test)
    print(f'kernel:{i}')
    print()
```
