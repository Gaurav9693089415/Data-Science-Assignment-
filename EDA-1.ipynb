```python
LB_mean=df['LB'].mean()
AC_mean=df['AC'].mean()
FM_mean=df['FM'].mean()
UC_mean=df['UC'].mean()
DL_mean=df['DL'].mean()
DS_mean=df['DS'].mean()
DP_mean=df['DP'].mean()
ASTV_mean=df['ASTV'].mean()
MSTV_mean=df['MSTV'].mean()
ALTV_mean=df['ALTV'].mean()
MLTV_mean=df['MLTV'].mean()
Width_mean=df['Width'].mean()
Tendency_mean=df['Tendency'].mean()
NSP_mean=df['NSP'].mean()
print(LB_mean,AC_mean,FM_mean,UC_mean,DL_mean,DS_mean,DP_mean,ASTV_mean,MSTV_mean,ALTV_mean,MLTV_mean,Width_mean,Tendency_mean,NSP_mean)
```

```python
df['LB'].fillna(LB_mean,inplace=True)
df['AC'].fillna(AC_mean,inplace=True)
df['DS'].fillna(DS_mean,inplace=True)
df['DP'].fillna(DP_mean,inplace=True)
df['MLTV'].fillna(MLTV_mean,inplace=True)
df['Width'].fillna(Width_mean,inplace=True)
df['Tendency'].fillna(Tendency_mean,inplace=True)
df['NSP'].fillna(NSP_mean,inplace=True)
```

## Detecting outliers


```python
col_nam=list(df.columns)
col_nam
```

```python
import matplotlib.pyplot as plt
```

```python
## visualisatoion of outliers by boxplot
fig=plt.figure(figsize=(20,10))
for i in col_nam:
    df.boxplot(column=col_nam)
    plt.show()
    
```

```python
## finding outliers manually
Q1=df.quantile(.25)
Q2=df.quantile(.50)
Q3=df.quantile(.75)
IQR=Q3-Q1
print('IQR  for each column:',IQR)
```

```python
upper_quant=Q3+1.5*IQR
print('upper_quantile:',upper_quant)
low_quant=Q1-1.5*IQR
print('lower quantile:',low_quant)
```
